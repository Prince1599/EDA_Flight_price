```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
# still these are string ,so we need to convert into numerical feature.
```

```python
df.head(2)
```

```python
# to convert string  into numerical feature using astype(int)
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
df.drop('Arrival_Time',axis=1  ,inplace=True)
```

```python
df.head(2)
```

```python
# to convert string  into numerical feature using astype(int)
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Hour',axis =1,inplace=True)
df.drop('Arrival_Minute',axis =1,inplace=True)
```

```python
df.info()
```

```python
df['Dept_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dept_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
# to convert string  into numerical feature using astype(int)
df['Dept_hour'] = df['Dept_hour'].astype(int)
df['Dept_min'] = df['Dept_min'].astype(int)
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head(2)
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Duration_hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
# converting NaN value into 0 
df['Duration_hour'] = df['Duration_hour'].fillna(0)
df['Duration_min'] = df['Duration_min'].fillna(0)
df.head(2)
```

```python
df.info()
```

```python
# Clean the 'Duration_hour' column by removing non-numeric characters
df['Duration_hour'] = df['Duration'].str.replace('[^\d]', '', regex=True)

# Clean the 'Duration_min' column by removing non-numeric characters
df['Duration_min'] = df['Duration'].str.replace('[^\d]', '', regex=True)

# Convert the cleaned columns to integers
df['Duration_hour'] = df['Duration_hour'].astype(int)
df['Duration_min'] = df['Duration_min'].astype(int)
df.drop('Duration',axis=1,inplace=True)
```
